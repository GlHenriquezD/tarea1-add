```python
dolar_2020_txt.columns=['Día', 'Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
nueva_linea_txt = pd.DataFrame({'Día':'1', 'Ene': 'NaN', 'Feb':'NaN', 'Mar':'NaN', 'Abr':852.03, 'May':'NaN', 'Jun':806.32, 'Jul':821.23, 'Ago':'NaN', 'Sep':776.46, 'Oct':788.15, 'Nov':'NaN', 'Dic':767.29}, index=[0])
dolar_2020_txt_nombres = pd.concat([nueva_linea_txt,dolar_2020_txt.loc[:]]).reset_index(drop=True)

dolar_2020_txt_nombres_rows_dropped = dolar_2020_txt_nombres.drop(dolar_2020_txt_nombres.index[[31]])

dolar_2020_txt_columns = dolar_2020_txt_nombres_rows_dropped.assign(Year=2020)

dolar_2020_txt_final = dolar_2020_txt_columns.astype('float64')
dolar_2020_txt_final
```

**DATAFRAME dolar_2020_regex**
